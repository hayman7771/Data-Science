## IMPORT LIBRARY (NUMPY)

```python
import numpy as np
```

## SPECIFY ARRAYS (a, b)

```python

A = np.array([[2, 1], [1, 3]])
b = np.array([1, 2])
```

## USE linalg MODULE TO FIND SOLUTION TO Ax = B

```python
x = np.linalg.solve(A, b)
print("Solution to Ax = b:", x)
```

## USE linalg TO FIND EIGENVALUES & EIGENVECTORS OF A

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```
