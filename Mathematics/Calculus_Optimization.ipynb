## IMPORT LIBRARIES

```python
import numpy as np
```

```python

def f(x: float) -> float:
    return (x - 3) ** 2 + 5
```

```python
def f_prime(x: float) -> float:
    return 2 * (x - 3)
```

```python
x = 0.0
lr = 0.1
```

```python
for i in range(30):
    grad = f_prime(x)
    x -= lr * grad
```

```python
print("Approximate minimizer:", x)
print("f(x) at minimum:", f(x))
```
