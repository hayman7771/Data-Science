## IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## DEFINE CLASSIFICATION METRICS

```python
X, labels_true = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=42)
```

## DEFINE ALGORITHM & SPECIFY CLUSTERING (for k-means)

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X)
```

## PLOT RESULTS:

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker="x", s=200)
plt.title("K-Means Clustering Example")
plt.tight_layout()
plt.show()
```
